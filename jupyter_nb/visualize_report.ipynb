```python
dpt_exp_names = ['NYU_norm_dpt', 'NYU_orig_dpt']
unsup_exp_names = ['NYU_norm_unsup', 'NYU_orig_unsup']
```

```python
df_dpt = []
for exp_name in dpt_exp_names:
    save_path = f'/mnt/cfs/sihaozhe/ss-dfd/exp/{exp_name}/result/eval.csv'
    df_exp = pd.read_csv(save_path, index_col=0)
    df_dpt.append(df_exp)
df_dpt = pd.concat(df_dpt)
```
